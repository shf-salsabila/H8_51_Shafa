```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'],
           axis=1, inplace=True)

df_can.head(2)
```

## Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%',
                           startangle=90) # 1.1f 1 angka di belakang koma dalam bentuk f/f log

plt.title('Immigration to Canada by Continent')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral',
              'lightskyblue', 'lightgreen', 'plum']

explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                           labels=None,
                           pctdistance=1.12,
                           colors=colors_List
                           explode=explode_list) # 1.1f 1 angka di belakang koma dalam bentuk f/f log

plt.title('Immigration to Canada by Continent', y=12)
plt.axis('equal')

plt.legend(labels=df_continents.index)

plt.show()
```

## Scatter Plot

```python
# gunanya untuk ngebandingin dua variabel. Kalo titik2nya naik, maka korelasinya positif, kalo salah satu variabelnya nambah, maka lainnya nambah.
# Kalo titik2nya naik turun maka korelasinya lemah.
# Untuk ngeliat 2 variabel ada hubungannya atau ga
```

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.head()
```

```python
df_tot.index =  map(int, df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.head()
```

```python
df_tot.columns = ['years', 'total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter',
           x='years',
           y='total',
           figsize=(10,6))
plt.title('Total Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
