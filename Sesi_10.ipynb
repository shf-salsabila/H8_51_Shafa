```python
%matplotlib inline

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True) # Styling
sns.set(rc={'figure.figsize':(5,5)}) # deafult ukuran canvas
```

```python
from scipy.stats import uniform

n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
data_uniform
```

```python
ax = sns.distplot(data_uniform,
                 bins=100,
                 kde=True,
                 color='skyblue',
                 hist_kws={'linewidth': 15, 'alpha': 1})

ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

## Confidence Interval

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset/heart.csv')
```

```python
df.head()
```

```python
df['Sex1'] = df.sex.replace({1: 'Male', 0: 'Female'})
```

```python
df.head()
```

```python
df.groupby('Sex1').agg({'chol': [np.mean, np.std, np.size]})
```

```python
mean_fe = 261.45
sd = 64.4
n = 312
z = 1.96 #95% confidence
```

```python
se = sd / np.sqrt(n)
```

```python
low_CI = mean_fe - z * se
hi_CI = mean_fe + z * se

low_CI, hi_CI
```

Dengan confidence level sebanyal 95%, saya percaya bahwa tingkat kolesterol wanita ada di antara 254.303 sampai 268.596

## Hypothesis Testing

```python
da = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv')
```

```python
df.head()
```

```python
females = da[
    da['RIAGENDR'] == 2
]

male = da[
    da['RIAGENDR'] == 1
]
```

```python
n1 = len(females)
mu1 = females['BMXBMI'].mean()
sd1 = females['BMXBMI'].std()

n1, mu1, sd1
```

```python
n2 = len(male)
mu2 = male['BMXBMI'].mean()
sd2 = male['BMXBMI'].std()

n2, mu2, sd2
```

```python
import statsmodels.api as sm
```

```python
sm.stats.ztest(
    females['BMXBMI'].dropna(),
    male['BMXBMI'].dropna()
)
```

p-value = 6.59e-10, lebih kecil daripada alpha yaitu 0.05 atau 5%.
Reject null hypothesis, BMI males sama dengan BMI female. Menerima alternative hypothesis yaitu BMI male tidak sama dengan BMI female.
